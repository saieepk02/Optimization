```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# -------- Non-linear Function --------
def f(x):
    return x[0]**2 + 4*np.sin(x[0])

# Store convergence values
history = []

def callback(x):
    history.append(f(x))

# Initial guess
x0 = [5]

# Nelder-Mead Optimization
result = minimize(f, x0, method='Nelder-Mead', callback=callback)

# Output
print("Optimal x:", result.x)
print("Minimum value:", result.fun)

# -------- Convergence Graph --------
plt.plot(history)
plt.title("Convergence of Nelder-Mead")
plt.xlabel("Iterations")
plt.ylabel("Function Value")
plt.grid()
plt.show()

#Nelder-Mead is a local optimization algorithm
#It works using a simplex method
#It does not require derivatives
#The graph shows how the function value decreases over iterations (convergence)
#Non-linear function → x² + sin(x)
#Nelder-Mead finds local minimum without using derivatives.”
```
