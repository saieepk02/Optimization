```python
import numpy as np
import matplotlib.pyplot as plt

# Minimization function
def loss(x):
    return (x + 2)**2

# Maximization function
def profit(x):
    return -(x - 3)**2 + 9

# Values
x = np.linspace(-10, 10, 100)

y1 = loss(x)
y2 = profit(x)

# Optimal points
min_x = -2
min_y = loss(min_x)

max_x = 3
max_y = profit(max_x)

# Plot
plt.plot(x, y1, label="Minimization (Loss)")
plt.plot(x, y2, label="Maximization (Profit)")

plt.scatter(min_x, min_y)
plt.text(min_x, min_y, "Min")

plt.scatter(max_x, max_y)
plt.text(max_x, max_y, "Max")

plt.title("Function Optimization")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.grid()

plt.show()

# Output
print("Minimum at x =", min_x, "Value =", min_y)
print("Maximum at x =", max_x, "Value =", max_y)

#👉 “This program shows minimization and maximization.”
#👉 “Minimum at x = -2, maximum at x = 3.”
#👉 “Used in optimization problems in Machine Learning.”
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function with multiple minima
def f(x):
    return x**2 + 10*np.sin(x)

# Generate values
x = np.linspace(-10, 10, 200)
y = f(x)

# Global minimum (lowest point)
g_x = x[np.argmin(y)]
g_y = f(g_x)

# Local minimum (approx value)
l_x = 3.8
l_y = f(l_x)

# Plot graph
plt.plot(x, y, label="Cost Function")

# Mark global minimum
plt.scatter(g_x, g_y)
plt.text(g_x, g_y, "Global Min")

# Mark local minimum
plt.scatter(l_x, l_y)
plt.text(l_x, l_y, "Local Min")

# Labels
plt.title("Cost Function Visualization")
plt.xlabel("x")
plt.ylabel("Cost")
plt.legend()
plt.grid()

plt.show()

#👉 “This function has multiple minima.”
#👉 “Global minimum is the lowest point overall.”
#👉 “Local minimum is lowest in a small region.”
#👉 “Graph helps visualize optimization landscape.”
```

```python
import numpy as np

# (a) Define function
def f(x):
    return (x - 3)**2 + 2   # minimum at x = 3

# (b) Random Search
random_x = np.random.uniform(-10, 10, 50)
random_y = f(random_x)

best_random_x = random_x[np.argmin(random_y)]
best_random_y = f(best_random_x)

# (c) Brute Force Search
x = np.linspace(-10, 10, 100)
y = f(x)

best_brute_x = x[np.argmin(y)]
best_brute_y = f(best_brute_x)

# Results
print("Random Search:")
print("Best x =", round(best_random_x,2), "Value =", round(best_random_y,2))

print("\nBrute Force:")
print("Best x =", round(best_brute_x,2), "Value =", round(best_brute_y,2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function
def f(x):
    return x**2 - 4*x + 3

# Range
x = np.linspace(-10, 10, 100)
y = f(x)

# -------- Brute Force --------
bf_x = x[np.argmin(y)]
bf_y = f(bf_x)

# -------- Random Search --------
rand_x = np.random.uniform(-10, 10, 30)   # 30 random points
rand_y = f(rand_x)

rs_x = rand_x[np.argmin(rand_y)]
rs_y = f(rs_x)

# -------- Plot --------
plt.plot(x, y, label="Function")

# Plot random points
plt.scatter(rand_x, rand_y)

# Mark best points
plt.scatter(bf_x, bf_y)
plt.text(bf_x, bf_y, "Brute Min")

plt.scatter(rs_x, rs_y)
plt.text(rs_x, rs_y, "Random Min")

plt.title("Random Search vs Brute Force")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

plt.show()

# Output
print("Brute Force Min:", bf_x, bf_y)
print("Random Search Min:", rs_x, rs_y)

#👉 “Brute force checks all values to find exact minimum.”
#👉 “Random search checks random points to find approximate minimum.”
#👉 “Random search is faster but less accurate.”
```

```python
import numpy as np

# Cost function
def f(x, y):
    return (x - 2)**2 + (y - 3)**2

# Define parameter ranges (grid)
x_range = np.arange(-5, 6, 1)
y_range = np.arange(-5, 6, 1)

best_x = 0
best_y = 0
min_value = float('inf')

# Grid Search
for x in x_range:
    for y in y_range:
        value = f(x, y)
        
        if value < min_value:
            min_value = value
            best_x = x
            best_y = y

# Output
print("Best x:", best_x)
print("Best y:", best_y)
print("Minimum value:", min_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function
def f(x, y):
    return (x - 2)**2 + (y - 3)**2

# Parameter ranges
x_range = np.arange(-5, 6, 1)
y_range = np.arange(-5, 6, 1)

best_x = 0
best_y = 0
min_value = float('inf')

all_x = []
all_y = []

# Grid Search
for x in x_range:
    for y in y_range:
        value = f(x, y)
        
        all_x.append(x)
        all_y.append(y)
        
        if value < min_value:
            min_value = value
            best_x = x
            best_y = y

# -------- Plot --------
plt.scatter(all_x, all_y, label="Grid Points")

# Best point
plt.scatter(best_x, best_y)
plt.text(best_x, best_y, "Best")

plt.title("Grid Search Optimization")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.show()

# Output
print("Best x:", best_x)
print("Best y:", best_y)
print("Minimum value:", min_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# -------- UNIVARIATE FUNCTION --------
x = sp.symbols('x')
f1 = x**2 - 4*x + 3

# Analytical gradient
grad1 = sp.diff(f1, x)
print("Univariate Gradient:", grad1)

# Numerical gradient
def func1(x):
    return x**2 - 4*x + 3

def num_grad1(x):
    h = 0.0001
    return (func1(x+h) - func1(x-h)) / (2*h)

print("Numerical Gradient at x=1:", num_grad1(1))

# -------- MULTIVARIATE FUNCTION --------
x, y = sp.symbols('x y')
f2 = x**2 + y**2

# Partial derivatives (gradient)
grad_x = sp.diff(f2, x)
grad_y = sp.diff(f2, y)

print("Multivariate Gradient:")
print("df/dx =", grad_x)
print("df/dy =", grad_y)

# -------- Visualization (Univariate only for simplicity) --------
x_vals = np.linspace(-5, 5, 100)
y_vals = func1(x_vals)

x0 = 1
y0 = func1(x0)
slope = num_grad1(x0)

tangent = y0 + slope * (x_vals - x0)

plt.plot(x_vals, y_vals, label="Function")
plt.plot(x_vals, tangent, '--', label="Gradient Line")

plt.scatter(x0, y0)
plt.text(x0, y0, "Point")

plt.title("Gradient Visualization")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# -------- Univariate --------
x = sp.symbols('x')
f = x**2 - 4*x + 3
print("Gradient:", sp.diff(f, x))

def func(x): return x**2 - 4*x + 3
def grad(x): return (func(x+0.001) - func(x-0.001))/0.002

print("Numerical:", grad(1))

# -------- Multivariate --------
x, y = sp.symbols('x y')
f2 = x**2 + y**2
print("df/dx:", sp.diff(f2, x))
print("df/dy:", sp.diff(f2, y))

# -------- Graph --------
xv = np.linspace(-5,5,100)
yv = func(xv)

x0 = 1
y0 = func(x0)
s = grad(x0)

plt.plot(xv, yv)
plt.plot(xv, y0 + s*(xv-x0), '--')  # tangent

plt.scatter(x0, y0)
plt.title("Gradient")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# -------- UNIVARIATE --------
x = sp.symbols('x')
f1 = x**3 + 2*x**2 + x

grad1 = sp.diff(f1, x)
print("Univariate Gradient:", grad1)

# For graph
def func1(x):
    return x**3 + 2*x**2 + x

# Numerical gradient
def grad_num(x):
    h = 0.001
    return (func1(x+h) - func1(x-h)) / (2*h)

print("Numerical Gradient at x=1:", grad_num(1))
xv = np.linspace(-5, 5, 100)
yv = func1(xv)

x0 = 1
y0 = func1(x0)
slope = grad_num(x0)

# -------- MULTIVARIATE --------
x, y = sp.symbols('x y')
f2 = x**2 + y**2 + x*y

df_dx = sp.diff(f2, x)
df_dy = sp.diff(f2, y)

print("df/dx:", df_dx)
print("df/dy:", df_dy)

# -------- PLOT --------
plt.plot(xv, yv, label="Univariate Function")

# Tangent (gradient)
plt.plot(xv, y0 + slope*(xv-x0), '--', label="Gradient Line")

plt.scatter(x0, y0)
plt.text(x0, y0, "Point")

plt.title("Gradient Visualization (Uni + Multi)")
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# -------- Generate Noisy Data --------
x = np.linspace(0, 10, 50)
y = 2*x + 3 + np.random.normal(0, 2, size=len(x))  # linear + noise

# -------- Linear Fit --------
coeff = np.polyfit(x, y, 1)
y_linear = np.polyval(coeff, x)

# -------- Non-linear Fit --------
def func(x, a, b):
    return a*np.exp(b*x)

params, _ = curve_fit(func, x, y, maxfev=10000)
y_nonlinear = func(x, *params)

# -------- Error (MSE) --------
mse_linear = np.mean((y - y_linear)**2)
mse_nonlinear = np.mean((y - y_nonlinear)**2)

print("Linear Error:", mse_linear)
print("Non-linear Error:", mse_nonlinear)

# -------- Plot --------
plt.scatter(x, y, label="Data")
plt.plot(x, y_linear, label="Linear Fit")
plt.plot(x, y_nonlinear, label="Non-linear Fit")

plt.title("Least Squares Curve Fitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function and gradient
def f(x):
    return x**2

def grad(x):
    return 2*x

# Iterations
n = 30
x0 = 5

# -------- Basic GD --------
x = x0
gd = []
for i in range(n):
    x = x - 0.1 * grad(x)
    gd.append(f(x))

# -------- Momentum --------
x = x0
v = 0
mom = []
for i in range(n):
    v = 0.9*v + 0.1*grad(x)
    x = x - v
    mom.append(f(x))

# -------- Adagrad --------
x = x0
G = 0
ada = []
for i in range(n):
    g = grad(x)
    G += g**2
    x = x - (0.5/np.sqrt(G+1e-8))*g
    ada.append(f(x))

# -------- RMSProp --------
x = x0
G = 0
rms = []
for i in range(n):
    g = grad(x)
    G = 0.9*G + 0.1*g**2
    x = x - (0.1/np.sqrt(G+1e-8))*g
    rms.append(f(x))

# -------- Adam --------
x = x0
m, v = 0, 0
adam = []
for i in range(n):
    g = grad(x)
    m = 0.9*m + 0.1*g
    v = 0.999*v + 0.001*g**2
    x = x - 0.1*m/(np.sqrt(v)+1e-8)
    adam.append(f(x))

# -------- Plot --------
plt.plot(gd, label="GD")
plt.plot(mom, label="Momentum")
plt.plot(ada, label="Adagrad")
plt.plot(rms, label="RMSProp")
plt.plot(adam, label="Adam")

plt.title("Optimizer Comparison")
plt.xlabel("Iterations")
plt.ylabel("Function Value")
plt.legend()
plt.grid()

plt.show()

#“All optimizers minimize same function.”
#👉 “Difference is in convergence speed.”
#👉 “Adam converges fastest in most cases.”
#👉 “Momentum reduces oscillations.”
```
