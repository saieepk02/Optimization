```python
import numpy as np
import matplotlib.pyplot as plt

# Function with multiple minima
def f(x):
    return x**2 + 10*np.sin(x)

# Generate values
x = np.linspace(-10, 10, 200)
y = f(x)

# Global minimum (lowest point)
g_x = x[np.argmin(y)]
g_y = f(g_x)

# Local minimum (approx value)
l_x = 3.8
l_y = f(l_x)

# Plot graph
plt.plot(x, y, label="Cost Function")

# Mark global minimum
plt.scatter(g_x, g_y)
plt.text(g_x, g_y, "Global Min")

# Mark local minimum
plt.scatter(l_x, l_y)
plt.text(l_x, l_y, "Local Min")

# Labels
plt.title("Cost Function Visualization")
plt.xlabel("x")
plt.ylabel("Cost")
plt.legend()
plt.grid()

plt.show()

#👉 “This function has multiple minima.”
#👉 “Global minimum is the lowest point overall.”
#👉 “Local minimum is lowest in a small region.”
#👉 “Graph helps visualize optimization landscape.”
```
