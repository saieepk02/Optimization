```python
import numpy as np

# Cost function
def f(x, y):
    return (x - 2)**2 + (y - 3)**2

# Define parameter ranges (grid)
x_range = np.arange(-5, 6, 1)
y_range = np.arange(-5, 6, 1)

best_x = 0
best_y = 0
min_value = float('inf')

# Grid Search
for x in x_range:
    for y in y_range:
        value = f(x, y)
        
        if value < min_value:
            min_value = value
            best_x = x
            best_y = y

# Output
print("Best x:", best_x)
print("Best y:", best_y)
print("Minimum value:", min_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function
def f(x, y):
    return (x - 2)**2 + (y - 3)**2

# Parameter ranges
x_range = np.arange(-5, 6, 1)
y_range = np.arange(-5, 6, 1)

best_x = 0
best_y = 0
min_value = float('inf')

all_x = []
all_y = []

# Grid Search
for x in x_range:
    for y in y_range:
        value = f(x, y)
        
        all_x.append(x)
        all_y.append(y)
        
        if value < min_value:
            min_value = value
            best_x = x
            best_y = y

# -------- Plot --------
plt.scatter(all_x, all_y, label="Grid Points")

# Best point
plt.scatter(best_x, best_y)
plt.text(best_x, best_y, "Best")

plt.title("Grid Search Optimization")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.show()

# Output
print("Best x:", best_x)
print("Best y:", best_y)
print("Minimum value:", min_value)
```
