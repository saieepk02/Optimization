```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# -------- UNIVARIATE FUNCTION --------
x = sp.symbols('x')
f1 = x**2 - 4*x + 3

# Analytical gradient
grad1 = sp.diff(f1, x)
print("Univariate Gradient:", grad1)

# Numerical gradient
def func1(x):
    return x**2 - 4*x + 3

def num_grad1(x):
    h = 0.0001
    return (func1(x+h) - func1(x-h)) / (2*h)

print("Numerical Gradient at x=1:", num_grad1(1))

# -------- MULTIVARIATE FUNCTION --------
x, y = sp.symbols('x y')
f2 = x**2 + y**2

# Partial derivatives (gradient)
grad_x = sp.diff(f2, x)
grad_y = sp.diff(f2, y)

print("Multivariate Gradient:")
print("df/dx =", grad_x)
print("df/dy =", grad_y)

# -------- Visualization (Univariate only for simplicity) --------
x_vals = np.linspace(-5, 5, 100)
y_vals = func1(x_vals)

x0 = 1
y0 = func1(x0)
slope = num_grad1(x0)

tangent = y0 + slope * (x_vals - x0)

plt.plot(x_vals, y_vals, label="Function")
plt.plot(x_vals, tangent, '--', label="Gradient Line")

plt.scatter(x0, y0)
plt.text(x0, y0, "Point")

plt.title("Gradient Visualization")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from mpl_toolkits.mplot3d import Axes3D

# -------- UNIVARIATE --------
def func1(x):
    return x**2 - 4*x + 3

def grad_num(x):
    h = 0.0001
    return (func1(x+h) - func1(x-h)) / (2*h)

x_vals = np.linspace(-5, 5, 100)
y_vals = func1(x_vals)

x0 = 1
y0 = func1(x0)
slope = grad_num(x0)

tangent = y0 + slope * (x_vals - x0)

# -------- MULTIVARIATE --------
def func2(x, y):
    return x**2 + y**2

x2 = np.linspace(-5, 5, 50)
y2 = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x2, y2)
Z = func2(X, Y)

# -------- PLOTS --------
plt.figure(figsize=(12,5))

# 1D Graph
plt.subplot(1,2,1)
plt.plot(x_vals, y_vals, label="Function")
plt.plot(x_vals, tangent, '--', label="Gradient Line")
plt.scatter(x0, y0)
plt.title("Univariate Gradient")
plt.legend()
plt.grid()

# 3D Graph
ax = plt.subplot(1,2,2, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title("Multivariate Surface")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()

#Univariate shows slope, multivariate shows surface and gradient direction.”
```
