```python
import numpy as np
import matplotlib.pyplot as plt

# Function
def f(x):
    return x**2 - 4*x + 3

# Range
x = np.linspace(-10, 10, 100)

y = f(x)

# -------- Brute Force --------
bf_x = x[np.argmin(y)]
bf_y = f(bf_x)

# -------- Random Search --------
rand_x = np.random.uniform(-10, 10, 30)   # 30 random points
rand_y = f(rand_x)

rs_x = rand_x[np.argmin(rand_y)]
rs_y = f(rs_x)

# -------- Plot --------
plt.plot(x, y, label="Function")

# Plot random points
plt.scatter(rand_x, rand_y)

# Mark best points
plt.scatter(bf_x, bf_y)
plt.text(bf_x, bf_y, "Brute Min")

plt.scatter(rs_x, rs_y)
plt.text(rs_x, rs_y, "Random Min")

plt.title("Random Search vs Brute Force")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()

plt.show()

# Output
print("Brute Force Min:", bf_x, bf_y)
print("Random Search Min:", rs_x, rs_y)

#👉 “Brute force checks all values to find exact minimum.”
#👉 “Random search checks random points to find approximate minimum.”
#👉 “Random search is faster but less accurate.”
```
