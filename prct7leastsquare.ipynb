```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# -------- Generate Noisy Data --------
x = np.linspace(0, 10, 50)
y = 2*x + 3 + np.random.normal(0, 2, size=len(x))  # linear + noise

# -------- Linear Fit --------
coeff = np.polyfit(x, y, 1)
y_linear = np.polyval(coeff, x)

# -------- Non-linear Fit --------
def func(x, a, b):
    return a*np.exp(b*x)

params, _ = curve_fit(func, x, y, maxfev=10000)
y_nonlinear = func(x, *params)

# -------- Error (MSE) --------
mse_linear = np.mean((y - y_linear)**2)
mse_nonlinear = np.mean((y - y_nonlinear)**2)

print("Linear Error:", mse_linear)
print("Non-linear Error:", mse_nonlinear)

# -------- Plot --------
plt.scatter(x, y, label="Data")
plt.plot(x, y_linear, label="Linear Fit")
plt.plot(x, y_nonlinear, label="Non-linear Fit")

plt.title("Least Squares Curve Fitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.show()

#Least Squares Method:Minimizes the sum of squared errors
#“This program generates noisy data and fits it using linear and non-linear models. 
#It uses least squares method to minimize error and compares both models using MSE. Linear fit performs better because the data is linear.”
```
