```python
import numpy as np
import matplotlib.pyplot as plt

# Minimization function
def loss(x):
    return (x + 2)**2

# Maximization function
def profit(x):
    return -(x - 3)**2 + 9

# Values
x = np.linspace(-10, 10, 100)

y1 = loss(x)
y2 = profit(x)

# Optimal points
min_x = -2
min_y = loss(min_x)

max_x = 3
max_y = profit(max_x)

# Plot
plt.plot(x, y1, label="Minimization (Loss)")
plt.plot(x, y2, label="Maximization (Profit)")

plt.scatter(min_x, min_y)
plt.text(min_x, min_y, "Min")

plt.scatter(max_x, max_y)
plt.text(max_x, max_y, "Max")

plt.title("Function Optimization")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.grid()

plt.show()

# Output
print("Minimum at x =", min_x, "Value =", min_y)
print("Maximum at x =", max_x, "Value =", max_y)

#👉 “This program shows minimization and maximization.”
#👉 “Minimum at x = -2, maximum at x = 3.”
#👉 “Used in optimization problems in Machine Learning.”
#This code plots two quadratic functions. One shows minimization with minimum at x = -2,
#and the other shows maximization with maximum at x = 3. It demonstrates optimization concepts used in machine learning
```
